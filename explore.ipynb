```python
response = request.urlopen("https://nomadlist.com/api/v2/list/places")
place = response.read().decode("utf-8")
place_dict = json.loads(place)
place_dict.keys()
```

```python
place_dict["result"][0].keys()
```

```python
place_dict["result"][0]["media"]["image"]["500"].split("/")[-1].split("500px")[0]
```

```python

# dict_cost = place_dict["result"][21]['cost']
def transform_cost_dict(dict_cost):
    for k,v in dict_cost.items():
        dict_cost[k] = v['USD']
    return dict_cost   
# transform_cost_dict(dict_cost)
```

```python
# dict_cost = place_dict["result"][15]['cost']
# dict_cost
```

```python
place_list = []
i = 0
for place in place_dict["result"]:
    new_dict = place["scores"]
    print(i)
    i +=1
    print(place['cost'])
    x = transform_cost_dict(place['cost'])
    print(x)
    new_dict.update(x)
    new_dict["city_name"] = place["media"]["image"]["500"].split("/")[-1].split("500px")[0].strip("-")
    place_list.append(new_dict)
```

```python
newdf = pd.DataFrame(place_list)
newdf.head()
```

```python
newdf['country'] = newdf['city_name'].apply(lambda x: get_country(x))
country_list = get_list("countries")
newdf['country_slug'] = newdf['country'].apply(lambda x: get_corect_country(x, country_list))
newdf['country_name'] = newdf['country_slug'].apply(lambda x: get_country_name(x))
```

```python
def get_mean_values_by(df,country, factor):
    agr = df.groupby(country)[factor].mean().round(2)
    return agr.to_dict()
```
